```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from collection import Collection
```

```python
try:
    collection
except NameError:
    collection = Collection()
```

```python
collection.import_file('dbs/final/final-cpv-fourth-with-sprache/auto.json', force=True)
df = collection.df_row_per_algorithm()
df['attribute_nr'] = df['attributes'].str.split(',').str.len()
df['attributes'] = df['attributes'].str.split(',').str[-1]
df['mcc_mean'] = df['mcc_mean']*100
df
```

```python
collection.import_file('dbs/final/final/auto.json', force=True)
df_all = collection.df_row_per_algorithm()
df_all['attribute_nr'] = df_all['attributes'].str.split(',').str.len()
df_all['attributes'] = df_all['attributes'].str.split(',').str[-1]
df_all['mcc_mean'] = df_all['mcc_mean']*100
df_all
```

# CPV first

```python
df_all.groupby(['attributes','algorithm']).mean().unstack().sort_values(('attribute_nr', 'decision_tree')).plot(y=['acc_mean', 'mcc_mean'], figsize=(12,8))
```

```python
from cycler import cycler
default_cycler = (cycler(color=[u'#1f77b4', u'#ff7f0e', u'#2ca02c', u'#d62728', u'#9467bd', u'#8c564b', u'#e377c2', u'#7f7f7f', u'#bcbd22']) +
                  cycler(linestyle=['-', ':', '-.', '-', ':', '-.', '-', ':', '-.']))
plt.rc('axes', prop_cycle=default_cycler)
#ax = df.groupby(['attributes','label']).mean().unstack().sort_values(('attribute_nr', 'Construction')).plot(figsize=(12,8), y=['acc_mean', 'mcc_mean'])
ax = df.groupby(['algorithm', 'attributes']).mean().reset_index().pivot('attributes', 'algorithm').sort_values(('attribute_nr', 'decision_tree')).plot(y=['acc_mean', 'mcc_mean', 'fn_rate_mean'], figsize=(12,8))
ax.legend(title="(Metric, Algorithm)", fontsize=14, bbox_to_anchor=(0.56, 0.4))
```
