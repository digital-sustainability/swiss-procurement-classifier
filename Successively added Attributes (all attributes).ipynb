```python
%matplotlib inline
import matplotlib.pyplot as plt
from collection import Collection
```

```python
try:
    collection
except NameError:
    collection = Collection()
```

```python
collection.import_file('dbs/final/cpv-fourth/auto.json', force=True)
```

```python
df = collection.df_row_per_algorithm()
df
```

```python
df['attributes'].str.split(",")
```

```python
df['attribute_nr'] = df['attributes'].str.split(',').str.len()
df['attributes'] = df['attributes'].str.split(',').str[-1]
df['mcc_mean'] = df['mcc_mean']*100
```

```python
df.sort_values('attribute_nr').groupby(['algorithm', 'anbieter']).plot(y=['acc_mean', 'mcc_mean', 'fn_rate_mean'], x='attributes')
```

```python
fig, ax = plt.subplots(figsize=(12,8))
df.sort_values('attribute_nr').groupby(['algorithm', 'anbieter']).mean()
#.plot(y=['acc_mean'], x='attributes', ax=ax)
```

```python
tmp_df = df.sort_values('attribute_nr').groupby(['attributes','anbieter']).mean().unstack().sort_values(('attribute_nr', 'Alpiq AG'))['acc_mean'].plot(figsize=(12,8))
```

# Plot for each Algorithm

```python
for algorithm, subdf in df.groupby(['algorithm']):
    subdf.groupby(['attributes']).mean().sort_values('attribute_nr').plot(y=['acc_mean', 'mcc_mean', 'fn_rate_mean'], title=algorithm, figsize=(12,8))
```

# Filtered dataframe with sample_size over threshold

```python
for algorithm, subdf in df[df['sample_size_mean']>=40].groupby(['algorithm']):
    subdf.groupby(['attributes']).mean().sort_values('attribute_nr').plot(y=['acc_mean', 'mcc_mean', 'fn_rate_mean'], title=algorithm, figsize=(12,8))
```

```python
df.groupby(['label', 'attributes']).mean().reset_index().pivot('attributes','label').sort_values(('attribute_nr','Construction')).plot(y=['acc_mean', 'mcc_mean', 'fn_rate_mean'], title=algorithm, figsize=(12,8))
```
