# change sample_size_mean

`sample_size_mean` is the number of test samples.
The number of train samples is 3 times as much

```python
df['sample_size_mean'] = df['sample_size_mean'].apply(lambda x: x*3)
```

# sample_size_mean to number of positives in DB

```python
tmpdf = df[df['sample_size_mean']<100].groupby('sample_size_mean').mean()
ax = tmpdf.plot(y=['acc_std','mcc_std','fn_std'], figsize=(12,8), fontsize=14)
ax.set_xlabel("Training Samples", fontsize=16)
ax.set_ylabel("Standard Deviation", fontsize=16)
```
